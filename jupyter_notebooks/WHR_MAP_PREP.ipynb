# Prepare geodata for the map

```python
# dependencies
import json
import requests
import pandas as pd
import numpy as np
```

# 2018

```python
whr_2018 = pd.read_csv('whr_2018.csv')

# Drop column 'Unnamed: 0'
whr_2018.drop('Unnamed: 0', axis=1, inplace=True)
whr_2018.head()
```

```python
whr_2018.info()
```

```python
# replace missing values with string 'n/a'
whr_2018 = whr_2018.fillna('n/a')
whr_2018.info()
```

```python
whr_2018.set_index('cca2', inplace=True)
whr_2018.head()
```

```python
whr_2018_dict = whr_2018.to_dict('index')

whr_2018_dict
```

```python
with open('country_data_2018.json', 'w') as file:
    json.dump(whr_2018_dict, file)
```

```python
with open('country_data_2018.json') as file:
    countries_2018 = json.load(file)
countries_2018
```

```python
# load country boundaries
with open('country_polygons.json') as file:
    polygons = json.load(file)
```

```python
features = []

# Prepare data for geoJson format
for item in polygons['polygon_features']:
    cca2 = item['properties']['cca2']
    for country in whr_2018_dict:
        if country.lower() == cca2:
            feature = {}
            feature['type'] = 'Feature'
            feature['properties'] = whr_2018_dict[country]
            feature['properties']['cca2'] = country
            feature['geometry'] = item['geometry']
            features.append(feature)
```

```python
geodata_2018 = {
    'type': 'FeatureCollection',
    'features': features
}
```

```python
with open('geodata_2018.json', 'w') as file:
    json.dump(geodata_2018, file)
```

# Produce years 2017 down to 2005 in similar fashion

```python
whr_2005 = pd.read_csv('whr_2005.csv')

# Drop column 'Unnamed: 0'
whr_2005.drop('Unnamed: 0', axis=1, inplace=True)
whr_2005.head()
```

```python
whr_2005.info()
```

```python
# replace missing values with string 'n/a'
whr_2005 = whr_2005.fillna('n/a')
whr_2005.info()
```

```python
whr_2005.set_index('cca2', inplace=True)
whr_2005.head()
```

```python
whr_2005_dict = whr_2005.to_dict('index')
whr_2005_dict
```

```python
features = []

# Prepare data for geoJson format
for item in polygons['polygon_features']:
    cca2 = item['properties']['cca2']
    for country in whr_2005_dict:
        try:
            if country.lower() == cca2:
                feature = {}
                feature['type'] = 'Feature'
                feature['properties'] = whr_2005_dict[country]
                feature['properties']['cca2'] = country
                feature['geometry'] = item['geometry']
                features.append(feature)
        except AttributeError:
            continue
```

```python
geodata_2005 = {
    'type': 'FeatureCollection',
    'features': features
}
```

```python
with open('geodata_2005.json', 'w') as file:
    json.dump(geodata_2005, file)
```
